### Import Libraries


```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Load the dataset

```python

df = pd.read_csv('data_date.csv')
```

#### Quick view

```python

print("First few rows of the dataset:")
print(df.head())
```

#### check distribution of status

```python

print("\nStatus Value Counts:")
print(df['Status'].value_counts())

print("\nStatus Percentage Distribution:")
print(df['Status'].value_counts(normalize=True) * 100)
```

# --- Visualization Part ---

#### 1. Countplot for Air Quality Status

```python

distribution_order = df['Status'].value_counts().index
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Status', order=distribution_order, palette='viridis')
plt.title('Air Quality Status Distribution')
plt.xlabel('Count')
plt.ylabel('Status')
plt.tight_layout()
plt.show()
```

#### 2. Histogram for AQI Values

```python

plt.figure(figsize=(8,6))
plt.hist(df['AQI Value'], bins=30, edgecolor='black', color='skyblue')
plt.title('Distribution of AQI Values')
plt.xlabel('AQI Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

#### 3. Top 10 Polluted Countries

```python

print("\nTop 10 Countries with Highest AQI Values:")
top10 = df.nlargest(10, 'AQI Value')
print(top10[['Country', 'AQI Value']])

plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='AQI Value', y='Country', palette='rocket')
plt.title('Top 10 Countries with Highest AQI Values')
plt.xlabel('AQI Value')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

#### 4. Boxplot of AQI Value by Status

```python

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Status', y='AQI Value', palette='coolwarm')
plt.title('AQI Value Distribution by Air Quality Status')
plt.xticks(rotation=45)
plt.xlabel('Air Quality Status')
plt.ylabel('AQI Value')
plt.tight_layout()
plt.show()
```
