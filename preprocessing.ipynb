```python
import pandas as pd
df = pd.read_csv('data_date.csv')
print(df.head())
```

```python
df['Status'].value_counts()
```

```python
df['Status'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
sns.boxplot(x='Status', y='AQI Value', data=df)
```

```python
df['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=80)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, y='Status', order=df['Status'].value_counts().index)
plt.title('Air Quality Status Distribution')
plt.xlabel('Count')
plt.ylabel('Status')
plt.show()
```
